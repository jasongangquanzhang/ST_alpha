```python
from ST_alpha_env import ST_alpha_env
from ShortTermalpha import ShortTermalpha
import numpy as np
import matplotlib.pyplot as plt
```

```python
short_term_alpha = ShortTermalpha(zeta=0.5, epsilon=0.002, eta=0.001)

env = ST_alpha_env(
    ShortTermalpha=short_term_alpha,
    nu=0,
    T=60,
    POV=0.2,
    Nq=20,
    S_0=30,
    X_0=0,
    q_0=1,
    Delta=0.01,
    varphi=0.01,
    phi=0,
    sigma=0.005,
)
```

```python
S,X,alpha,q,r=env.Simulate()
```

```python
which_plot = 5
```

```python
plt.figure(figsize=(12, 6))
plt.plot(S[which_plot].numpy(), alpha=0.3)
plt.title('S over Time')
plt.xlabel('Time Step')
plt.ylabel('S')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(alpha[which_plot].numpy(), alpha=0.3)
plt.axhline(alpha[which_plot].mean().item(), color='red', linestyle='--', label='Mean Alpha')
plt.legend()
plt.title('Alpha over Time')
plt.xlabel('Time Step')
plt.ylabel('Alpha')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(r[which_plot].numpy(), alpha=0.3)
plt.title('Reward over Time')
plt.xlabel('Time Step')
plt.ylabel('Reward')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(q[which_plot].numpy()*S[which_plot].numpy()+X[which_plot].numpy(), alpha=0.3)
plt.title('Reward over Time')
plt.xlabel('Time Step')
plt.ylabel('Reward')
plt.show()
```
