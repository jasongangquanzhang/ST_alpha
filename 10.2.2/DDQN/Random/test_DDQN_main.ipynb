```python
from env_standard_random import StandardEnv
from DDQN_standard_random import DDQNAgent

env = StandardEnv(
    T = 300,
    I_max = 20,
    lambd = 50/300,
    Delta = 0.01,
    phi = 0.000001,
    varphi = 0.001,
    sigma = 0.001,
    gamma = 0.999,
    S_0 = 20,
    X_0 = 0,
    q_0 = 0
)

agent = DDQNAgent(
    env=env, state_size = 4, action_size = 4, seed=42, lr=1e-3,
    weight_decay=0.01, sched_step_size=50/1, sched_gamma = 0.99,
    tau=1e-3, update_every=1, sample_size=1_000, capacity=1_000_000, name='trial',
    n_nodes=16, n_hidden_layers=6
)
```

```python
agent.train(n_iter=3_000, n_plot=500, eps_start=1.0, eps_end=0.01, eps_decay=0.9985, batch_size=500)
# agent.train(n_iter=5_000, n_plot=500, eps_start=1.0, eps_end=0.01, eps_decay=0.999, batch_size=1)
```
