```python
from env_standard_random import StandardEnv
from DDQN_standard_random import DDQNAgent

env = StandardEnv(
    T = 300,
    I_max = 20,
    lambd = 50/300,
    Delta = 0.01,
    phi = 1e-6,
    varphi = 7e-4,
    sigma = 0.001,
    gamma = 1,
    S_0 = 20,
    X_0 = 0,
    q_0 = 0
)

agent = DDQNAgent(
    env=env, state_size = 2, action_size = 4, seed=42, lr=1e-3,
    weight_decay=0.01, sched_step_size=500/1, sched_gamma = 0.99,
    tau=1e-3, update_every=1, sample_size=6_000, capacity=1_000_000, name='trial',
    n_nodes=32, n_hidden_layers=4     
)
```

```python
agent.train(n_iter=5_000, n_plot=500, eps_start=1.0, eps_end=0.01, eps_decay=0.999, batch_size=512)
# agent.train(n_iter=5_000, n_plot=500, eps_start=1.0, eps_end=0.01, eps_decay=0.999, batch_size=1)
```
